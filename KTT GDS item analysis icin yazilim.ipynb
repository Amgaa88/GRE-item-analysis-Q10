```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

```python
#creating dataframe
df=pd.read_csv('Oes-6th-Numerical.csv')
```

```python
df
```

```python
def rDf():
#response columns
    merged1=[]
    for i in range(1,11):
        response_cols=['Question-'+str(i)]
        merged=merged1.append(response_cols)
        merged2=list(itertools.chain(*merged1))
    #response dataframe
    response_df=df[merged2]
    response_df=response_df.applymap(lambda x:1 if x==1 else 0)
```

```python
#Item difficulty -p coefficient
a=response_df.mean()
def iDif(): 
    for i in a:
        if i>=0.81 and i<=1.0:
            print(str('%.2f' % i)+'-Very easy')
        elif  i>=0.61 and i<=0.80:
            print(str('%.2f' % i)+"-Easy")
        elif i>=0.41 and i<=0.60:
            print(str('%.2f' % i)+'-Medium difficulty')
        elif i>=0.21 and i<=0.40:
            print(str('%.2f' % i)+'-Difficult')
        else:
            print(str('%.2f' % i)+'-Very difficult') 
```

```python
#Item discrimination-r coefficient
def rCoef():
    quartiles=response_df.sum(axis=1).quantile([0.25, 0.75])
    #Split the students into upper and lower quartiles based on total score
    upper_quartile=df[response_df.sum(axis=1)>=quartiles.iloc[1]]
    lower_quartile=df[response_df.sum(axis=1)<=quartiles.iloc[0]]

    prop_correct_upper = upper_quartile.mean(numeric_only=True)
    prop_correct_lower = lower_quartile.mean(numeric_only=True)
    #Calculate the item discrimination index
    item_discrimination = prop_correct_upper - prop_correct_lower
```

```python
b=item_discrimination
def Discr(): 
    for r in b:
        if r>=0.81 and r<=1.0:
            print(str('%.2f' % r)+'-Very good')
        elif  r>=0.61 and r<=0.80:
            print(str('%.2f' % i)+"-Good")
        elif r>=0.41 and r<=0.60:
            print(str('%.2f' % r)+'-Medium')
        elif r>=0.21 and r<=0.40:
            print(str('%.2f' % r)+'-Poor')
        else:
            print(str('%.2f' % r)+'-Very poor') 
```

```python
#Combined Function
def combined_func():
    result1=rDf()
    result2=iDif()
    result3=rCoef()
    result4=Discr()
    return(result1,result2,result3,result4)
   
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Calculate the mean and standard deviation of the sum of correct answers
scores = response_df.sum(axis=1)
mean = np.mean(scores)
std_dev = np.std(scores)

# Generate x-values and calculate y-values for the bell curve
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
y = np.exp(-np.power(x - mean, 2) / (2*std_dev**2)) / (std_dev*np.sqrt(2*np.pi))

# Plot the bell curve
plt.plot(x, y)
plt.xlabel('  ')
plt.ylabel(' ')
plt.title('  ')

# Add vertical lines for mean and mean plus/minus one standard deviation
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1)
plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=1)
plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=1)

# Add legend for the lines
plt.legend(['Bell Curve', f' ({mean:.2f})', f'  ({std_dev:.2f})'])

plt.show()
```
